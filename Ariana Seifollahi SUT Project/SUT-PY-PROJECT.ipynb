        به نمودار رسم شده در قسمت قبل، نمودار روند تغییرات مرتبه اول(خطی) را اضافه کنید.(برای پیدا کردن ضرایب رگرسیون می‌توانید از تابع polyfit در کتابخانه numpy استفاده کنید.)

```python
df_sorted = df.sort_values(by='Area')
df_cleaned = df_sorted[['Area', 'Price(USD)']].dropna()
df_cleaned['Area'] = (df_cleaned['Area'] - df_cleaned['Area'].mean()) / df_cleaned['Area'].std()
slope, intercept = np.polyfit(df_cleaned['Area'], df_cleaned['Price(USD)'], 1, rcond=1e-10)
linear_fit = np.polyval([slope, intercept], df_cleaned['Area'])
print(f'Slope = {slope}, Intercept = {intercept}')
plt.figure(figsize=(12, 8))
plt.plot(df_cleaned['Area'], df_cleaned['Price(USD)'], marker='o', label='Original Data')
plt.plot(df_cleaned['Area'], linear_fit, label='Linear Fit')
plt.title('Sorted Broken Line Chart with Linear Fit of Price(USD) by Area')
plt.xlabel('Area (Standardized)')
plt.ylabel('Price(USD)')
plt.legend()
plt.xlim([df_cleaned['Area'].min(), df_cleaned['Area'].max() + 0.1])
plt.ylim([df_cleaned['Price(USD)'].min(), df_cleaned['Price(USD)'].max()])
plt.grid(True)
plt.show()
```

```python
df_sorted = df.sort_values(by='Area')
df_cleaned = df_sorted[['Area', 'Price(USD)']].dropna()
df_cleaned['Area'] = (df_cleaned['Area'] - df_cleaned['Area'].mean()) / df_cleaned['Area'].std()
slope, intercept = np.polyfit(df_cleaned['Area'], df_cleaned['Price(USD)'], 1, rcond=1e-10)
linear_fit = np.polyval([slope, intercept], df_cleaned['Area'])
print(f'Slope = {slope}, Intercept = {intercept}')
plt.figure(figsize=(12, 8))
plt.plot(df_cleaned['Area'], df_cleaned['Price(USD)'], marker='o', label='Original Data')
plt.plot(df_cleaned['Area'], linear_fit, linestyle='--', label='Linear Fit')
plt.title('Sorted Broken Line Chart with Linear Fit of Price(USD) by Area')
plt.xlabel('Area (Standardized)')
plt.ylabel('Price(USD)')
plt.legend()
plt.xlim([df_cleaned['Area'].min(), df_cleaned['Area'].max() + 0.1])
plt.ylim([df_cleaned['Price(USD)'].min(), df_cleaned['Price(USD)'].max()])
plt.grid(True)
plt.show()
```

نمودارهای میله‌ای یکی از ابزارهای قوی در تجزیه و تحلیل داده‌های گروه‌بندی شده هستند. این نمودارها به شما امکان می‌دهند تا مقدار یک متغیر در گروه‌های مختلف را با یکدیگر مقایسه کنید و با طراحی بهینه و استفاده از رنگ‌ها و اندازه‌های مختلف، این نمودارها می‌توانند اطلاعات زیادی را به شما منتقل کنند.
استفاده از این نوع نمودارها برای تحلیل آماری داده‌ها، نمایش توزیع فراوانی‌ها، مقایسه گروه‌ها و موارد دیگر در تحلیل داده‌ها بسیار کمک‌کننده است. 
میله‌ها در نمودار به طور عمودی یا افقی می‌توانند نمایش داده شوند.
